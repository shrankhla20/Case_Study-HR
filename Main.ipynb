```python
df = df.drop(['EmployeeCount','EmployeeNumber','StandardHours'], axis =1)
```

```python
df.columns
```

```python
num_col = list(df.describe().columns)
num_col
```

```python
cat_col = list(set(df.columns).difference(num_col))
```

```python
cat_col
```

```python
cat_col.remove('Attrition')
```

```python
df['Attrition'].unique()
```

```python
df['Attrition'] = df['Attrition'].map({'Yes':0,'No':1})
```

```python
df['Attrition'].unique()
```

```python
df_cat = pd.get_dummies(df[cat_col])
df_cat
```

```python
y = df['Attrition']
```

```python
X = pd.concat([df[num_col],df_cat], axis =1)
```

```python
X.columns
```

```python
X.info()
```

# DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.3,random_state=1)
```

```python
model = DecisionTreeClassifier(random_state =1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
metrics.plot_confusion_matrix(model, X_test, y_test)
```
